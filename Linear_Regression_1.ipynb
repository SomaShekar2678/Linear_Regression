```python
plt.figure(figsize=(5,5)) # siz of my graph
slope,intercept=np.polyfit(e['actual'],e['predicted'],1) # help me to get alues of slope n intercept
best_fit_line=slope*e['actual']+intercept # mx+c
plt.scatter(e['actual'],e['predicted'],color='crimson')
plt.plot(e['actual'],best_fit_line,color='black')
plt.xlabel("true values")
plt.ylabel("prediction")
```
